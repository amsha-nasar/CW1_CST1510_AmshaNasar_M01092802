### Activity 14: Counting Different Sizes per Clothing Item

Create a Python program that reads data from the `clothing_stock.txt` file and counts how many different sizes of each clothing item are available. Store this in a dictionary and print it.

**Steps:**
1. Read the file and extract both clothing items and sizes.
2. Count how many different sizes exist for each item.
3. Store the results in a dictionary and print it.
<hr>

`````python

item_sizes = {}  # Key: clothing item, Value: set of sizes


with open("clothing_stock.txt", "r") as f:
    for line in f:
        line = line.strip()
        if line == "":
            continue  # Skip empty lines
        
        # Split line into item and size
        parts = line.split()
        if len(parts) != 2:
            continue  # Skip malformed lines
        
        item, size = parts[0], parts[1]
        
        # Step 3: Add size to the set corresponding to the item
        if item not in item_sizes:
            item_sizes[item] = set()
        item_sizes[item].add(size)

# Step 4: Count sizes and print
size_count = {item: len(sizes) for item, sizes in item_sizes.items()}

print("Number of different sizes per clothing item:")
for item, count in size_count.items():
    print(f"{item}: {count} sizes")
`````

### Activity 15: Managing Study Groups

You are responsible for managing a study group for a course. Each student can be part of multiple study groups, and you need to track which groups a student is enrolled in. Students can join or leave study groups. What data structure would you choose to manage this efficiently, and why?

**Example data:**

```python
{"John": ["Math", "Physics"], "Emily": ["Physics", "Biology"], "Alex": ["Math"]}

Steps:

1. Choose a data structure that allows quick access to the groups each student belongs to.
2. Write a program that lets students join and leave study groups.
3. Implement a function to display all groups a student is part of.
<hr>

`````python

students = {
    "John": {"Math", "Physics"},
    "Emily": {"Physics", "Biology"},
    "Alex": {"Math"}
}

# Function to join a group
def join_group(student, group):
    if student not in students:
        students[student] = set()  # Add new student if not exists
    students[student].add(group)
    print(f"{student} has joined {group}.")

# Function to leave a group
def leave_group(student, group):
    if student in students and group in students[student]:
        students[student].remove(group)
        print(f"{student} has left {group}.")
    else:
        print(f"{student} is not in {group} or student does not exist.")

# Function to display all groups of a student
def display_groups(student):
    if student in students:
        print(f"{student} is in groups: {', '.join(students[student])}")
    else:
        print(f"{student} does not exist.")


join_group("John", "Biology")
leave_group("Emily", "Physics")
display_groups("John")
display_groups("Emily")
display_groups("Alex")
`````

### Activity 16: Managing and Shuffling Playlists

You are developing a music player that lets users create a playlist and shuffle it. Users should be able to add songs, remove songs, and shuffle the playlist randomly. What data structure would you use to represent the playlist, and why?

**Example data:**

```python
["Song A", "Song B", "Song C", "Song D"]
````

Steps:

1. Choose a data structure that allows easy addition, removal, and shuffling of songs.
2. Write a program that simulates playlist operations (add, remove, shuffle).
3. Let users shuffle the playlist and display the new order.
<hr>

`````python
import random

# Initial playlist
playlist = ["Song A", "Song B", "Song C", "Song D"]

# Function to display playlist
def display_playlist():
    print("Current Playlist:")
    for i, song in enumerate(playlist, start=1):
        print(f"{i}. {song}")
    print()

# Function to add a song
def add_song(song):
    playlist.append(song)
    print(f"Added '{song}' to the playlist.\n")

# Function to remove a song
def remove_song(song):
    if song in playlist:
        playlist.remove(song)
        print(f"Removed '{song}' from the playlist.\n")
    else:
        print(f"'{song}' not found in the playlist.\n")

# Function to shuffle the playlist
def shuffle_playlist():
    random.shuffle(playlist)
    print("Playlist shuffled!\n")

# --- Example usage ---
display_playlist()
add_song("Song E")
remove_song("Song B")
shuffle_playlist()
display_playlist()
`````
