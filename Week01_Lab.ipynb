### Stretch 3. Population Projection (5 Years)

Assume current population is 312032486 and 1 year = 365 days.
Rates: birth every 7s, death every 13s, immigrant every 45s.
Compute projected population for each of the next 5 years using integer division (`//`).

**Example Output:**
```
Year 1: 315,xxx,xxx
...
Year 5: ...
```


*Optional hint:* Compute yearly net change once; then add each year.


````python
#M01092802
#Amsha Nasar
curr=312032486             # initial value for population 
birth=(3600*24*365)//7    # to calculate total number of births 
death=(3600*24*365)//13   # to calculate total number of deaths
imm=(3600*24*365)//45     # to calculate total number of immigrants coming in 

curr2=curr+birth+imm-death   # calculate the population for the next year
net_change=curr2-curr        # calculate net change by subtracting the first year and second year 

print (f'The net change each year is {net_change}')
print("year 1: ", curr2)
curr2+=net_change         # use of augmented assignment 
print("year 2: ", curr2)
curr2+=net_change       # add net change to previous year's population value 
print("year 3: ", curr2)
curr2+=net_change
print("year 4: ", curr2)
curr2+=net_change
print("year 5: ", curr2)




````

### Stretch 4. Payroll Statement

Ask for: employee name, hours worked, hourly rate, federal tax rate, state tax rate.
Print a formatted payroll slip with gross pay, each deduction, and net pay (2 d.p.).

**Example Output:**
```
Employee: Smith
Gross pay: £97.50
Federal withholding: £19.50
State withholding: £8.78
Net pay: £69.22
```


````python
#M01092802
#Amsha Nasar

name=input("enter employee name: ")
hours=float(input('enter your hours worked: '))
hourly_rate=float(input('enter your hourly rate %: '))   #take hourly rate as input 
state=float(input('enter your fedral tax rate%: '))     #take state rate as input 
fed=float(input('enter your state tax rate: '))    #take fedral rate as input 

gross_pay=hours*hourly_rate       
fed_tax=gross_pay*fed/100               #money taken away as fedral tax 
state_tax=gross_pay*state/100           #money taken away as state tax 
net_pay=gross_pay-fed_tax-state_tax    # take home pay is gross salary minus all the taxes

print(f"Employee: {name} \nGross pay:$ {gross_pay:.2f}")

print(f"Fedral Withholding:$ {fed_tax:.2f} \nState Withholding:$ {state_tax:.2f} \nNet Pay:$ {net_pay:.2f}")  # use 2f to round to 2 decimal place
````
